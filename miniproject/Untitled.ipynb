```python
from bs4 import BeautifulSoup as soup
page_soup=soup(response.text,'html.parser')
```

```python
page_soup
```

```python
main_box=page_soup.find_all('div',{'class':'_1-2Iqu row'})
```

```python
data_list=[]
for box in main_box:
    temp_dict={}
    temp_dict['title']=box.find('div',{'class':'_3wU53n'}).text.rstrip()
    temp_dict['price']=box.find('div',{'class':'_1vC4OE _2rQ-NK'}).text.rstrip()
    temp_dict['rating']=box.find('div',{'class':'hGSR34'}).text.rstrip()
    data_list.append(temp_dict)
```

```python
import pandas as pd
df=pd.DataFrame(data_list)
```

```python
df.to_csv("samsung_price_list.csv")
```
